```python
# !pip install keras
# !pip install tensorflow
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_full = pd.read_excel('./Final_Project/ODIR-5K/ODIR-5K/data.xlsx', sheet_name='Sheet1')
df_full
```

```python
num = df_full['Right-Fundus'].apply(lambda x:x.split('_')[0])
df_full.insert(7,'num',num)
df_full
```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(df_full,test_size =0.3,random_state=0)
x_train = train.iloc[:,0:8]
y_train = train.iloc[:,8:]
x_test = test.iloc[:,0:8]
y_test = test.iloc[:,8:]
```

```python
x_test
```

```python
rightimg_train = []
for i in range(x_train.shape[0]):
    img = image.load_img('./Final_Project/ODIR-5K/ODIR-5K/Training Images/'+x_train['num'].iloc[i]+'_right.jpg',target_size=(100,100))
    img = image.img_to_array(img)
    img = img/255
    rightimg_train.append(img)
x_train_r = np.array(rightimg_train)
x_train_r.shape
```

```python
leftimg_train = []
for i in range(x_train.shape[0]):
    img = image.load_img('./Final_Project/ODIR-5K/ODIR-5K/Training Images/'+x_train['num'].iloc[i]+'_left.jpg',target_size=(100,100))
    img = image.img_to_array(img)
    img = img/255
    leftimg_train.append(img)
x_train_l = np.array(leftimg_train)
x_train_l.shape
```

```python
rightimg_test = []
for i in range(x_test.shape[0]):
    img = image.load_img('./Final_Project/ODIR-5K/ODIR-5K/Training Images/'+x_test['num'].iloc[i]+'_right.jpg',target_size=(100,100))
    img = image.img_to_array(img)
    img = img/255
    rightimg_test.append(img)
x_test_r = np.array(rightimg_test)
x_test_r.shape
```

```python
leftimg_test = []
for i in range(x_test.shape[0]):
    img = image.load_img('./Final_Project/ODIR-5K/ODIR-5K/Training Images/'+x_test['num'].iloc[i]+'_left.jpg',target_size=(100,100))
    img = image.img_to_array(img)
    img = img/255
    leftimg_test.append(img)
x_test_l = np.array(leftimg_test)
x_test_l.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
```

```python
model = Sequential()
model.add(Conv2D(64, kernel_size=3, activation='relu', input_shape=(100,100,3)))
model.add(MaxPooling2D(pool_size=(2, 2)))  
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(8, activation='sigmoid'))
```

```python
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train_r,y_train, epochs=10, batch_size=64)
```

```python
y_predict = model.predict(x_test_r)
for i in y_predict:
    for j in range(len(i)):
        if i[j] == max(i):
            i[j] = 1
        else:
            i[j] = 0
count=0
for i in range(len(y_predict)):
    if np.where(y_predict[i] == 1) == np.where(y_test[i] == 1):
        count+=1
count/len(y_predict)
```

```python
y_predict = model.predict(x_test_r)
print(y_predict[0:5])
y_test.head(5)
```

```python
score = model.evaluate(x_test_r, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
